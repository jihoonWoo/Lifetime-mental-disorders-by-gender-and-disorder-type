## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/ABS-National-Survey-of-Mental-Health-and-Wellbeing/generated_data/'
input_file = 'transformed_data.xlsx'
```

```python
df = pd.read_excel(input_path + input_file, usecols=[1, 2, 3, 4, 5, 6, 7], sheet_name='lifetime')
df
```

The above data was generated from the National Survey of Mental Health and Wellbeing: Summary of Results 2007 in ABS [2]. It contains proportion of lifetime mental disorders by gender and disorder types in 2007. According to the transformed data, 45.5% of Australian reported experiencing a mental ill-health (disorder) at some point in their lifetime where male proportion size is around 5 percents greater than female's.

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
male = tuple(df.iloc[:-1, 4])
female = tuple(df.iloc[:-1, 5])

ind = np.arange(len(male))  # the x locations for the groups
width = 0.6  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/4, male, width/2,
                label='Male', color='#2B46C0')
rects2 = ax.bar(ind + width/4, female, width/2,
                label='Female', color='#C0392B')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Mental disorder types', fontsize=22)
ax.set_ylabel("Percents", fontsize=22)
ax.set_title('Australians mental ill-health in 2007', fontsize=28)
ax.set_xticks(ind)
ax.set_xticklabels(('Anxiety', 'Affective', 'Substance Use'), fontsize=18)
ax.legend(fontsize=12)
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')

fig.set_size_inches(16, 9)
fig.tight_layout()

for index, value in enumerate(male):
    plt.text(index - 0.2, value, str(round(value, 2)), fontsize=18)
for index, value in enumerate(female):
    plt.text(index + 0.1, value, str(round(value, 2)), fontsize=18)
    
plt.savefig('img/lifetime mental disorder by gender and disorder type.pdf', bbox_inches='tight')
plt.show()
```

Lifetime mental disorders can be analysed into three different categories: “anxiety disorders”, “affective disorders” and “substance use disorders” and also by gender. Women tend to report more  anxiety disorders than men who report more substance use disorders. 

## Insight

**32% of females in Australians (16-85 years) reported experiencing anxiety at some point in their lifetime in 2007.**
